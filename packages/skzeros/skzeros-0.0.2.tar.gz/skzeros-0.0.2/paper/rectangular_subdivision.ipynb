```python
import matplotlib.pyplot as plt
import numpy as np

from skzeros._domain import Rectangle, _subdivide_domain

plt.style.use("style.mplstyle")
```

```python
def f(z):
    return z**50 + z**12 - 5 * np.sin(20 * z) * np.cos(12 * z) - 1


def f_z(z):
    return (
        50 * z**49
        + 12 * z**11
        + 60 * np.sin(12 * z) * np.sin(20 * z)
        - 100 * np.cos(12 * z) * np.cos(20 * z)
    )
```

```python
r1 = Rectangle(complex(-20, -20), complex(20, 20))
leafs1 = _subdivide_domain(r1, f, f_z, 50.1)
```

```python
r2 = Rectangle(complex(-20, -20), complex(20, 20))
leafs2 = _subdivide_domain(r2, f, f_z, 25.1, maxiter=100)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(4, 3))
r1.plot(ax1)
r2.plot(ax2)
ax1.set_ylabel(r"$\Im(z)$")
for ax in (ax1, ax2):
    ax.set_xlabel(r"$\Re(z)$")
    ax.set_xlim([-21, 21])
    ax.set_ylim([-21, 21])
    ax.set_aspect("equal", "box")
plt.tight_layout()
plt.savefig("figures/example_subdivision.pdf")
```
