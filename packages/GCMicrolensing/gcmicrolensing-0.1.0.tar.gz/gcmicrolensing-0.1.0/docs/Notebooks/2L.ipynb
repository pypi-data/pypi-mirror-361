```python
import math

import matplotlib.pyplot as plt
import numpy as np
import VBMicrolensing
```

```python
VBM = VBMicrolensing.VBMicrolensing()

s = 0.9  # Separation between the lenses
q = 0.1  # Mass ratio
u0 = 0.0  # Impact parameter with respect to center of mass
alpha = 1.0  # Angle of the source trajectory
rho = 0.01  # Source radius
tE = 20.0  # Einstein time in days
t0 = 0  # Time of closest approach to center of mass
```

```python
# Array of parameters. Note that s, q, rho and tE are in log-scale
pr = [math.log(s), math.log(q), u0, alpha, math.log(rho), math.log(tE), t0]

t = np.linspace(t0 - tE, t0 + tE, 100)  # Array of times

magnifications, y1, y2 = VBM.BinaryLightCurve(
    pr, t
)  # Calculation of binary-lens light curve

plt.plot(t, magnifications)
```

```python
caustics = VBM.Caustics(s, q)
for cau in caustics:
    plt.plot(cau[0], cau[1])
plt.plot(y1, y2)
```

```python
criticalcurves = VBM.Criticalcurves(s, q)

# plot
fig = plt.figure(figsize=(5, 5))
for crit in criticalcurves:
    plt.plot(crit[0], crit[1], "k")
```

```python
import VBMicrolensing
from matplotlib import pyplot as plt

VBM = VBMicrolensing.VBMicrolensing()  # Instance to VBMicroLensing

# Parameters of our binary lens
s = 0.6
# separation between the two lenses
q = 0.1
# mass ratio

caustics = VBM.Caustics(s, q)

# plot
fig = plt.figure(figsize=(5, 5))
for cau in caustics:
    plt.plot(cau[0], cau[1], "k")
```

```python
criticalcurves = VBM.Criticalcurves(s, q)

# plot
fig = plt.figure(figsize=(5, 5))
for crit in criticalcurves:
    plt.plot(crit[0], crit[1], "k")
```

```python
VBM = VBMicrolensing.VBMicrolensing()

s = 0.8  # separation between the two lenses
q = 0.1  # mass ratio
y1 = 0.01  # y1 is the source coordinate along the axis parallel to the line joining the two lenses
y2 = 0.01  # y2 is the source coordinate orthogonal to the first one
rho = 0.01  # Source radius in Einstein radii

VBM.astrometry = True  # We want astrometry

Mag = VBM.BinaryMag2(
    s, q, y1, y2, rho
)  # Call to the BinaryMag2 function with these parameters
print(f"Binary lens Magnification = {Mag}\n")  # Output should be 18.28....
print(
    f"\nCentroid shift = ({VBM.astrox1 - y1},{VBM.astrox2 - y2})\n"
)  # Output should be (-0.164...,-0.074...)
```
