```python
# Load dataset relative to the script's location using __file__ and os.path
import os
current_dir = os.path.dirname(os.path.abspath(''))
data_path = os.path.join(current_dir, '../inputdata/data/historical_emissions/historical_emissions_image_regions.csv')

emissions = pd.read_csv(data_path).set_index("Year")
population = pd.read_csv(
    os.path.join(current_dir, '../inputdata/data/historical_population/historical_population.csv')
).set_index("Year")
```

```python
years = global_emissions.index
discount_factor = pd.Series(
    np.exp(-0.03 * (2020 - years)),
    index=years,
)
```
