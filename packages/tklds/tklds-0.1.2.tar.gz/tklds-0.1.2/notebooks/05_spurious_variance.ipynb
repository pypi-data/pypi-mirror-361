```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
from tklds.interface.generators import generate_lds_rvs, get_sequence_enum
from tklds.interface.processes import spurious_variance
```

```python
n = 1023
n_skip = 1
d_max = 21021
```

```python
sequence_enum = get_sequence_enum()
u = generate_lds_rvs(sequence_enum.TKRG_A_AP5, n=n, d=d_max)[n_skip:]
spv1 = spurious_variance(u)
```

```python
u = generate_lds_rvs(sequence_enum.NEW_JOE_KUO, n=n, d=d_max)[n_skip:]
spv2 = spurious_variance(u)
```

```python
sf, fs = 1.5,15 
fig,ax = plt.subplots(1,figsize=(sf*4,sf*3))

ax.semilogx(spv1, label="tkrg-a-ap5")
ax.semilogx(spv2, label="new-joe-kuo")

ax.axhline(0.0, linestyle="--", alpha=0.3, color="k")
ax.set_xlabel("Number of dimensions, d",fontsize=fs)
ax.set_ylabel("Spurious variance",fontsize=fs)

ax.set_title(f"Spurious variance\n{n=}, {d_max=}")
ax.legend(loc="upper left",fontsize=fs)
```
