```python
import numpy as np
import scipy
from scipy.special import erfinv
from matplotlib import pyplot as plt

from tklds.interface.generators import generate_lds_rvs, get_sequence_enum
from tklds.interface.processes import multivariate_brownian_motion
```

```python
num_steps = 200
num_dim = 2
```

```python
a_cov_diag = np.array([[0.1, 0.0],
                     [0.0, 1.0]])

a_cov_non_diag = np.array([[1, 1.0],
                         [1.0, 2]]) 

a_mu = np.zeros((1,num_dim))
a_x0 = np.zeros((1, num_dim))
```

```python
sequence_enum = get_sequence_enum()
a_u = generate_lds_rvs(sequence=sequence_enum.TKRG_A_AP5, n = num_steps, d = num_dim, skip=1)
```

```python
x = multivariate_brownian_motion(a_x0=a_x0, a_mu=a_mu, a_cov=a_cov_diag, a_u=a_u)
print(f"done, x.shape: {x.shape}")
```

```python
sf, fs = 1.2, 15
fig, ax = plt.subplots(1,figsize=(sf*4,sf*3))

for i in range(num_dim):
    ax.plot(x.T[i,:], label = f"path{i}")

ax.legend(bbox_to_anchor=(1,1))

ax.set_title("Sample paths for diagonal\n covariance matrix", fontsize=fs)
ax.set_ylabel("Value", fontsize=fs)
ax.set_xlabel("Time step", fontsize=fs)
```

```python
x = multivariate_brownian_motion(a_x0=a_x0, a_mu=a_mu, a_cov=a_cov_non_diag, a_u=a_u)
```

```python
sf, fs = 1.2, 15
fig, ax = plt.subplots(1,figsize=(sf*4,sf*3))

for i in range(num_dim):
    ax.plot(x.T[i,:], label = f"path{i}")

ax.legend(bbox_to_anchor=(1,1))

ax.set_title("Sample paths for non-diagonal\n covariance matrix", fontsize=fs)
ax.set_ylabel("Value", fontsize=fs)
ax.set_xlabel("Time step", fontsize=fs)
```
